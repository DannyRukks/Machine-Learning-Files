### Which of these elements had the highest sum in 2018?

```python
DT1 = DT[["Element", "Y2018"]]
```

```python
EleGRP = DT1.groupby("Element")
```

```python
EleGRP.sum().nlargest(1, "Y2018")
```

### Which of these elements had the 3rd lowest sum in 2018?

```python
EleGRP.sum().nsmallest(3, "Y2018")
```
